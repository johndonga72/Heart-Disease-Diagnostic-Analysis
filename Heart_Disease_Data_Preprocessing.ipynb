# Data Transformation

```python
# Imputation of erroneous data in 'ca' and 'thal' columns.

# We calculate the median and mode of ca and thal columns to be imputed in place of -100000.
# From below calculation, we come to know that median and mode for both 'ca' and 'thal' are 0.0 and 3.0
# So we will replace the -100000 with 0.0 in 'ca' and 3.0 in 'thal'
```

```python
heart_data.median()
```

```python
heart_data.mode()
```

```python
# Imputing 0 in 'ca' where value = -100000

ca_median = int(heart_data['ca'].median())
heart_data.loc[heart_data.ca == -100000, 'ca'] = np.nan
heart_data.fillna(ca_median,inplace=True)
```

```python
heart_data['ca']
```

```python
# Imputing 3 in 'thal' where value = -100000

thal_median = int(heart_data['thal'].median())
heart_data.loc[heart_data.thal == -100000, 'thal'] = np.nan
heart_data.fillna(thal_median,inplace=True)
```

```python
heart_data['thal']
```

```python
# Converting the numeric columns to categorical 

heart_data = heart_data.astype({"sex":'category', 
                                "cp":'category',
                                "fbs":'category',
                                "restecg":'category', 
                                "exang":'category',
                                "slope":'category',
                                "ca":'category', 
                                "thal":'category',
                                "num":'category'})
```

```python
# Detecting outliers in 'chol' columns and treating them by imputation

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Boxplot before outlier traeatment

plt.subplots(figsize=(15,10))
plt.boxplot(data=heart_data, x='chol')
```

```python
chol_mean = heart_data['chol'].mean()
chol_mean
```

```python
chol_outliers = []
def detect_outliers_iqr(data):
    data = sorted(data)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
 
    IQR = Q3-Q1
    lwr_bound = Q1-(1.5*IQR)
    upr_bound = Q3+(1.5*IQR)
   
    for i in data: 
        if (i < lwr_bound or i > upr_bound):
            chol_outliers.append(i)
    return chol_outliers

chol_outliers = detect_outliers_iqr(heart_data['chol'])
print("Outliers from IQR method for chol column: ", chol_outliers)
```

```python
for i in chol_outliers:
    heart_data['chol'] = np.where(heart_data['chol'] == i, heart_data['chol'].mean(), heart_data['chol'])
```

```python
# Boxplot after outlier traeatment

plt.subplots(figsize=(15,10))
plt.boxplot(data=heart_data, x='chol')
```
